<span style="font-size:80%">[문제6]<br><br></span>
```Python에서 메모리 관리는 어떻게 이루어지는지 설명해주세요```

<span style="color:red">답변</span>

Reference Counting 래퍼런스카운팅

객체 내부에 객체가 참조되는 횟수를 기록하고 객체가 더이상 참조되지 않아 참조횟수가 0이되면 객체의 메모리는 자동으로 해제된다.

이러한 래퍼런스 카운팅은 순환참조를 해결할 수 없다.
파이썬은 이를 해결하기 위해 가비지 컬렉션을 지원한다.
가비지컬렉션은 순환참조가 발생할 경우에만 동작하며, 메모리 누수를 방지한다.

* 순환참조는 두개 이상의 객체가 서로를 참조하는 상황을 말한다.
이러한 상황이 발생하면, 객체가 더이상 사용되지 않더라도 참조계수는 0이되지 않아 메모리 누수가 발생하게 된다.

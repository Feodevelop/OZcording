[문제1]

Python의 변수 할당 개념을 다른 언어와 비교해 설명해주세요


파이썬의 변수란 하나의 값을 저장할 수 있는 공간을 말한다.

파이썬, C++, 자바 등 많은 언어들은 연산자 = 를 활용해서 변수를 만든다.
C++와 같은 언어는 메모리에 변수가 저장되는 방식으로 관리된다. 새로운 변수에 선언된 변수를 할당하면 새로운 메모리에 동일한 객체가 복사되는 방식으로 저장된다. 그래서 동일 메모리의 변수를 가리키고 싶다면 C++은 포인터를 사용한다.

파이썬의 변수는 많은 포인터들로 구성되어 있으며, 변수 대입(=)은 새로운 메모리 할당이 아니라 객체에 대한 참조이다.
